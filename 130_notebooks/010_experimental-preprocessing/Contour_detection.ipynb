```python
%pylab inline
import cv2
import pathlib
```

```python
im = cv2.imread('img/test.jpg')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(contours)
```

```python
imshow(imgray, cmap="gray")
```

```python
im_contours = cv2.drawContours(im, contours,-1,(0,255,0),3)
```

```python
imshow(im_contours, cmap="gray")
```

```python
x,y,w,h = cv2.boundingRect(contours[0])
```

```python
im2 = cv2.rectangle(imgray, (x,y),(x+w,y+h),(255, 255, 255), 5)
```

```python
imshow(im2, cmap="gray")
```
